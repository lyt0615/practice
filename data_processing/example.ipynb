## 1 用[numpy](https://www.runoob.com/numpy/numpy-tutorial.html)数组表示数据

````python
import numpy as np

# np.array接受一个列表或者元组，构成一个numpy数组
a = [1,2,3]
b = (0.1,0.2,0.3)
c = [1]
d = [a,b]
e = [d,d]
print(np.array(a))
print(np.array(b))
print(np.array(c))
print(np.array(d)) # 这是一个2*3的二维数组，也就是数学里的矩阵
print(np.array(e)) # 这是一个2*2*3的三维数组，以此类推可以创建更高维度的数组
````
